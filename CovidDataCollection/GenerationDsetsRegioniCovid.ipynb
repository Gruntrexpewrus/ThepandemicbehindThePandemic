```python
df = pd.read_csv(r"C:\Users\leona\Desktop\Covid\dpc-covid19-ita-regioni.csv")
df.drop([ 'stato', 'note', 'lat', 'long', 'codice_regione'], axis = 1, inplace = True)
df.fillna(0, inplace = True)
df['data'] = df['data'].str.slice(0,10)
df
```

```python
regions = df['denominazione_regione'].unique()
for i in regions:
    temp = df.loc[df['denominazione_regione'] == i]
    temp.sort_values(by=['data'])
    name = i+'.csv'
    temp.to_csv(name, index=False)
    #print(temp)
```

```python
temp.head()
```

```python
regions_csvs = [i+'.csv' for i in regions]
regions_csvs
```

```python
tosubtract = ['tamponi', 'casi_testati', 'deceduti' ]
h = 0
for doc in regions_csvs:
    temp = pd.read_csv(doc)
    count_row = temp.shape[0] 
    temp['nuovi_tamponi'] = temp['tamponi'][0]
    temp['nuovi_casi_testati'] = temp['casi_testati'][0]
    temp['nuovi_deceduti'] =  temp['deceduti'][0]
    for i in range(1, count_row):
        for j in tosubtract:
           # print(temp[j][i]-temp[j][i-1])
            temp.at[i,'nuovi_'+j] = temp[j][i]-temp[j][i-1]
    temp.to_csv('DataCovidNEW/new'+doc, index = False)
```

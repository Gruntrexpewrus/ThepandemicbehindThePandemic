```python
def f(df):
    df.drop([ 'stato'], axis = 1, inplace = True)
    df.fillna(0, inplace = True)
    df['data'] = df['data'].str.slice(0,10)
    tosubtract = ['tamponi', 'casi_testati', 'deceduti' ]
    h = 0
    count_row = df.shape[0] 
    df['nuovi_tamponi'] = df['tamponi'][0]
    df['nuovi_casi_testati'] = df['casi_testati'][0]
    df['nuovi_deceduti'] =  df['deceduti'][0]
    for i in range(1, count_row):
            for j in tosubtract:
               # print(temp[j][i]-temp[j][i-1])
                df.at[i,'nuovi_'+j] = df[j][i]-df[j][i-1]
    df.to_csv('DataCovidNEW/NATIONALDATACOVIDDAYBYDAY.csv', index = False)
    return df
```
